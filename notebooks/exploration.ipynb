## 1. Core Lorentz Transformations

```python
from simulations.lorentz import gamma, boost_event, spacetime_interval, classify_interval

beta = 0.6
g = gamma(beta)
print(f"β = {beta}  →  γ = {g:.4f}")
print(f"Time dilation factor:    t = {g:.4f} τ")
print(f"Length contraction ratio: L = L₀ / {g:.4f} = {1/g:.4f} L₀")

# Transform a single event
event = (5.0, 3.0)   # (ct, x) in S
ct_p, x_p = boost_event(*event, beta)
print(f"\nEvent in S:  (ct={event[0]}, x={event[1]})")
print(f"Event in S': (ct'={ct_p:.4f}, x'={x_p:.4f})")
```

## 2. Spacetime Interval

```python
pairs = [
    ((0,0), (3,1),  'Timelike future'),
    ((0,0), (1,4),  'Spacelike'),
    ((0,0), (2,2),  'Lightlike'),
]

print(f"{'Description':<22} {'s²':>8}  Type")
print('-'*42)
for (p1, p2, label) in pairs:
    s2 = spacetime_interval(*p1, *p2)
    print(f"{label:<22} {s2:>8.2f}  {classify_interval(s2)}")
```
